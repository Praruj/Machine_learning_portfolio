# Tic tac Toe using MIN MAX

# 1. What is Intelligence?

Imagine someone asks:

> Should I cross the road?

A person does **not** answer immediately.

Instead, they mentally simulate the future.

```
Cross Road
     ↓
Car Comes
     ↓
Accident
```

or

```
Wait
   ↓
Car Passes
   ↓
Safe
```

This ability to **predict possible futures before acting** is one form of intelligence.

Games work exactly the same way.

Instead of acting immediately, we imagine future moves.

# 2. What is a Game?

A game can be viewed as a sequence of situations.

```
Current Situation
        ↓
Possible Move
        ↓
New Situation
        ↓
Possible Move
        ↓
New Situation
```

Computer scientists call each situation a **state**.

For Tic Tac Toe, one board configuration is one state.

Example:

```
X | O |
---------
| X |
---------
| | O
```

This board is a state.

If X plays in an empty square, we obtain another state.

---

# 3. State Representation

The computer stores the board as a list of lists.

```python
board = [
    [X, O, EMPTY],
    [EMPTY, X, EMPTY],
    [EMPTY, EMPTY, O]
]
```

Each move transforms one state into another.

Mathematically,

```
State + Action → New State
```

In Python:

```python
new_board = result(board, action)
```

The original board should never be modified directly.

Instead:

```python
copy.deepcopy(board)
```

creates a completely new board.

---

# 4. State Space

How many Tic Tac Toe boards are possible?

Not just one.

Thousands.

```
Board 1

Board 2

Board 3

...

Board N
```

The collection of **every possible board** is called the **State Space**.

The goal of Minimax is to search this state space intelligently.

---

# 5. Search Trees

Every move creates new possibilities.

```
Current Board
      |
 -------------------
 |       |        |
Move1  Move2   Move3
```

Each move creates another board.

Those boards create more moves.

```
               Root
            /    |    \
          A      B      C
        / | \   / | \   / | \
```

This structure is called a **Game Tree**.

Each node represents a board.

Each edge represents a move.

---

# 6. Terminal States

Eventually the game ends.

Possible endings:

- X wins
- O wins
- Draw

These final boards are called **Terminal States**.

They are the leaves of the tree.

```
          Root
        /      \
     Node      Node
     /  \      /  \
  Leaf Leaf  Leaf Leaf
```

Leaves have no children.

---

# 7. Utility Function

Computers work better with numbers than words.

Instead of

```
X Wins
```

use

```
+1
```

Instead of

```
Draw
```

use

```
0
```

Instead of

```
O Wins
```

use

```
-1
```

This mapping is called the **Utility Function**.

Example:

```python
def utility(board):

    if winner(board) == X:
        return 1

    elif winner(board) == O:
        return -1

    return 0
```

# 8. Rational Players

Assume X wants to win.

Suppose X has these possible outcomes.

```
Move A → +1

Move B → -1

Move C → 0
```

Which move should X choose?

Obviously

```
+1
```

X always wants the **largest score**.

Therefore,

**X is the Maximizer.**

---

Now consider O.

Possible outcomes:

```
+1

0

-1
```

O wants X to lose.

Therefore O chooses

```
-1
```

O always chooses the **smallest score**.

Therefore,

**O is the Minimizer.**

This is where the name **Minimax** comes from.

---

# 9. The Main Idea of Minimax

Suppose X has two possible moves.

```
              X
            /   \
           A     B
```

If X chooses A,

O will respond.

```
          A
        /   \
      -1     0
```

Since O is intelligent,

O chooses

```
-1
```

because it hurts X the most.

Therefore

```
A = -1
```

Now consider B.

```
        B
      /   \
     1     0
```

O chooses

```
0
```

because it is smaller than 1.

Therefore

```
B = 0
```

Now X compares

```
A = -1

B = 0
```

Since X maximizes,

```
max(-1,0)=0
```

So X chooses B.

# 10. Why Recursion?

Suppose we ask

> What is the best move from this board?

To answer,

we imagine one move.

Now we get another board.

Again we ask

> What is the best move from this board?

It is exactly the same question.

```
Current Board
      ↓
Future Board
      ↓
Future Board
      ↓
Future Board
```

The same problem repeats.

Whenever a problem contains a smaller version of itself,

**recursion is usually the right solution.**

---

# 11. Mapping the Idea to Python

## maxValue()

```python
def maxValue(board):

    if terminal(board):
        return utility(board)

    v = float("-inf")

    for action in actions(board):

        v = max(v,
                minValue(result(board, action)))

    return v
```

Meaning:

- If the game is over, return its score.
- Otherwise, try every possible move.
- Let the opponent respond.
- Keep the highest score.

---

## minValue()

```python
def minValue(board):

    if terminal(board):
        return utility(board)

    v = float("inf")

    for action in actions(board):

        v = min(v,
                maxValue(result(board, action)))

    return v
```

Meaning:

- If the game is over, return the score.
- Try every move.
- Let X respond.
- Keep the smallest score.

# Key Concepts

| Concept | Meaning |
|----------|---------|
| State | Current board configuration |
| Action | A legal move |
| Result | New board after a move |
| State Space | All possible board configurations |
| Game Tree | Tree of all possible moves |
| Terminal State | Final board where the game ends |
| Utility | Numerical value of a terminal state |
| Maximizer | Player trying to maximize the utility |
| Minimizer | Player trying to minimize the utility |
| Minimax | Algorithm that assumes both players play optimally |
| Recursion | Solving a problem by solving smaller versions of itself |

Imagine two perfect players talking.

```

"I will choose the move that hurts you the most."

↓
```

# Core Principle

> **Minimax is not about finding a winning move immediately. It is about choosing the move that guarantees the best possible outcome, assuming the opponent also plays perfectly.**


**code**


````python

import copy


X = "X"
O = "O"
EMPTY = None


def initial_state():
    """
    Returns starting state of the board.
    """
    return [
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY]
    ]


def player(board):
    """
    Returns player who has the next turn on a board.
    """

    count_x = sum(row.count(X) for row in board)
    count_o = sum(row.count(O) for row in board)

    if count_x > count_o:
        return O
    else:
        return X


def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """

    possible_actions = []

    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                possible_actions.append((i, j))

    return possible_actions


def result(board, action):
    """
    Returns the board that results from making move (i, j) on the board.
    """

    copied_board = copy.deepcopy(board)

    i, j = action

    current_player = player(board)

    copied_board[i][j] = current_player

    return copied_board


def winner(board):
    """
    Returns the winner of the game, if there is one.
    """

    # Check rows
    for row in board:
        if row.count(X) == 3:
            return X

        if row.count(O) == 3:
            return O


    # Check columns
    for j in range(3):
        column = []

        for i in range(3):
            column.append(board[i][j])

        if column.count(X) == 3:
            return X

        if column.count(O) == 3:
            return O


    # Check diagonals

    diagonal1 = [
        board[0][0],
        board[1][1],
        board[2][2]
    ]

    diagonal2 = [
        board[0][2],
        board[1][1],
        board[2][0]
    ]


    if diagonal1.count(X) == 3:
        return X

    if diagonal1.count(O) == 3:
        return O


    if diagonal2.count(X) == 3:
        return X

    if diagonal2.count(O) == 3:
        return O


    return None



def terminal(board):
    """
    Returns True if game is over, False otherwise.
    """

    if winner(board) is not None:
        return True

    if len(actions(board)) == 0:
        return True

    return False



def utility(board):
    """
    Returns 1 if X has won,
    -1 if O has won,
    0 otherwise.
    """

    winner_player = winner(board)

    if winner_player == X:
        return 1

    elif winner_player == O:
        return -1

    return 0



def minimax(board):
    """
    Returns the optimal action for the current player.
    """

    if terminal(board):
        return None


    current_player = player(board)


    if current_player == X:

        best_value = float("-inf")
        best_action = None

        for action in actions(board):

            new_board = result(board, action)

            value = min_value(new_board)

            if value > best_value:
                best_value = value
                best_action = action

        return best_action



    else:

        best_value = float("inf")
        best_action = None

        for action in actions(board):

            new_board = result(board, action)

            value = max_value(new_board)

            if value < best_value:
                best_value = value
                best_action = action

        return best_action



def max_value(board):

    if terminal(board):
        return utility(board)


    value = float("-inf")


    for action in actions(board):

        new_board = result(board, action)

        value = max(
            value,
            min_value(new_board)
        )


    return value



def min_value(board):

    if terminal(board):
        return utility(board)


    value = float("inf")


    for action in actions(board):

        new_board = result(board, action)

        value = min(
            value,
            max_value(new_board)
        )


    return value
````
