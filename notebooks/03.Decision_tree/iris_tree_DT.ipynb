```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
X, y = iris.data, iris.target
from sklearn.model_selection import train_test_split
# Split data: 60% train, 20% val, 20% test
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=0, stratify=y_temp)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.25, random_state=0, stratify=y_temp)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(iris.data,iris.target)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
```

## Training model A with pre-pruning: 

```python
# Pre pruning
clf_pre = DecisionTreeClassifier(
    max_depth=3,          # limit depth
    min_samples_split=5,  # min samples needed to split
    min_samples_leaf=2,   # min samples in a leaf
    max_leaf_nodes=10,    # max number of leaves
    random_state=0
)
clf_pre.fit(iris.data,iris.target)
```

### Pre Pruning
1. “Stop going down after 3 questions.” 
2. Don’t split a node if it has fewer than 5 samples.”
3. “Every final leaf must have at least 2 samples.”
4. Good: fast, simple tree.
5. Risk: we might stop too early and underfit (tree is too simple).

## Training Model B with post-pruning : 

```python
# post pruning
from sklearn.tree import DecisionTreeClassifier

clf_post = DecisionTreeClassifier(ccp_alpha=0.01, random_state=0)
clf_post.fit(iris.data,iris.target)
```

### Post-pruning : First let the tree grow big, then cut off weak branches.
- We can see what a full tree would look like and then simplify it.
- Cost: a bit more computation.


```python

best_model = max([classifier, clf_pre, clf_post], key=lambda m: m.score(X_val, y_val))
best_model.score(X_test,y_test)
```

```python

new_flower = [[5.1, 3.5, 1.4, 0.2]] 
prediction = best_model.predict(new_flower)
print("This flower is:", iris.target_names[prediction[0]]) 
```
