# Color Scale

 - Mathematical way of organizing and representing colors so that computers can understand them. 

Most common color spaces are:


1. **RGB/BGR**
    * Matplotlib use RGB, Opencv uses BGR, each pixel is mix of three values ranging from 0 to 255.
    * It is terrbile for color detection, if shadow falls over a red apple, red, green, blue values all change drastically. 

2. **HSV (human detection standard)**

    HSV splits image data into entirely different categories.
    * `HUE` :  actual color (eg: red, orange..) maps colors in a circle rom O $degree$ to 300 $degree$ (scaled to 0-179 in OpenCV).
    * `Saturation` : The intensity or purity of color (how vibrant or gray it is).
    * `Value`: The brightness of color (how light or dark is it).
    - used for color tracking and filtering. 

3. **Grayscale**

    * It throws away all color info, reducing image to a single channel representing brightness.
    * Pixels range from 0 (pure black) to 255 (pure white)
    * used for preprocessing, Edge/contour detection and speed.  Algorithms like Canny edge detection and thresholding require grayscale images to work.


```TEXT
Need to display an image?         ➔ Use BGR/RGB
Need to track a specific color?   ➔ Convert to HSV (cv.cvtColor(img, cv.COLOR_BGR2HSV))
Need to find shapes or edges?    ➔ Convert to Gray (cv.cvtColor(img, cv.COLOR_BGR2GRAY))

````python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


img=cv.imread("image/apple.jpg")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

#BGR to grayscle
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)

# BGR to HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

#BGR to L*a*b
lab=cv.cvtColor(img,cv.COLOR_BGR2Lab)
lab_for_matplotlib = cv.cvtColor(lab, cv.COLOR_BGR2RGB)

plt.imshow(lab_for_matplotlib)
plt.axis("off")
plt.show()
````
