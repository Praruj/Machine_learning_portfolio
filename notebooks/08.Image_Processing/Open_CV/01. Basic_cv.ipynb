### Essential Functions in cv
1.  Converting in gray scale 

    ```Python
    gray=cv2.cvtColor(image,cv.COLOR_BGR2GRAY)
    ```

2. Blur Image
    ```Python
    blur=cv2.GaussianBlur(img,(3,3),cv2.BORDR_DEFAULT)  #kernel size is always odd
    cv2.imshow("Blur",blur)
    ```

3. Edge cascade
    ```Python
    edge=cv.Canny(img,125,175)
    cv2.imshow("canny",edge)
    ```

4. Dilate Image 
    ```python
    dilated=cv.dilate(canny,(3,3).iterations=1)

    ```

5. Eroding 

```python
    eroded=cv2.erode(dilated,(3,3),iterations=1)
```

6. Resize
```Python
    cv2.resize(img,(500,500), interpolation=cv2.INTER_CUBIC)
```

````python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("../image/apple.jpg")

plt.subplot(2,2,1)
canny=cv2.Canny(img,125,175)
plt.imshow(canny)

plt.subplot(2,2,2)
blur=cv2.GaussianBlur(canny,(29,29),cv2.BORDER_DEFAULT)
plt.imshow(blur)


plt.subplot(2,2,3)
dilated=cv2.dilate(canny,(3,3),iterations=1)
plt.imshow(dilated)

plt.subplot(2,2,4)
eroded=cv2.erode(dilated,(3,3), iterations=1)
plt.imshow(eroded)
plt.axis("off")
plt.show()  
````

* Dilation make bright regions (edges/foreground) thicker or grow shapes in a binary image.

    - Input: Binary or grayscale image (e.g., result of Canny).
    - Kernel: Small matrix (like 3×3 ones) that defines how pixels grow.
    - Iterations: Number of times the growth is applied.

* Erosion : If dilation grows white regions, then erosion shrinks white regions.
    - Removes small white noise
    - Thins edges
    - Shrinks foreground objects
