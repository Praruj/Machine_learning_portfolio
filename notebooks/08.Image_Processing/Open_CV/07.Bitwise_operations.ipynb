# Bitwise Operators in Opencv

```TEXT
An image is just a massive grid of numbers—pixels. And what are those numbers made of Bits. Zeros and ones. By comparing the bits of two images, we can create masks, hide complex backgrounds, and isolate exactly what we want with incredible speed.

The fundamental operators:
```
* Rule of thumb
- 0 = Black (off)
- 255=White(On)

1. Bitwise AND (`cv2.bitwise_and`)
<br>

I will only turn a pixel ON if both Image A and Image B have that pixel turned ON.
- True only if both input are true. 
- if we overlay whiteshape (255) onto an image, only the paet of image is coverd by white shape will remain everything will become black (0).


2.  Bitwise OR (`cv2.bitwise_or`)
<br>

I will turn a pixel ON if either Image A or Image B (or both) have that pixel turned ON.
- True if at least one input is True
- it combines the white parts of both images together


3. Bitwise XOR (`cv2.bitwise_xor`)
<br>

The NOT operator is the contrarian. It doesn't compare two images; it just flips a single image.
- inverts the bits ( Turn 0 to 1 and vice versa)
- It photographic inverts image, white becomes black and black becomes white. 

```TEXT
Why Use Them?
Masking & Region of Interest (ROI): Acting like a digital stencil to isolate specific objects (e.g., green screens, lane detection) while ignoring the background.

Watermarking & Overlays: Seamlessly inserting non-rectangular shapes or logos onto a background.

Hardware Speed: Operates directly on binary data, replacing slow Python loops with instant, pixel-parallel math.

````python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

blank=np.zeros((400,400),dtype="uint8")
rectangle=cv.rectangle(blank.copy(),(30,30),(370,370),255,-1)
circle=cv.circle(blank.copy(),(200,200),200,255,-1)


plt.subplot(1,2,1)
plt.imshow(rectangle, cmap="gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(circle, cmap="gray")
plt.axis("off")
plt.show()
````

```TEXT
By default, Matplotlib is designed to make data stand out, so it applies that colorful "Viridis" map. By explicitly adding cmap="gray", we are telling Matplotlib: "Do not try to be fancy. Just map 0 to black, 255 to white, and everything else to shades of gray."

**Bitwise_AND ---> Finds intersecting region**

````python
# Bitwise AND

bitwise_and=cv.bitwise_and(rectangle,circle)
plt.imshow(bitwise_and,cmap="gray")
plt.axis("off")
plt.show()
````

**Bitwise_OR ---> Finds Non intersecting region and intersecting regions**

````python
# Bitwise OR

bitwise_or=cv.bitwise_or(rectangle,circle)
plt.imshow(bitwise_or,cmap="gray")
plt.axis("off")
plt.show()
````

**Bitwise_AND ---> Finds non intersecting region**

````python
# Bitwise XOR

bitwise_xor=cv.bitwise_xor(rectangle,circle)
plt.imshow(bitwise_xor,cmap="gray")
plt.axis("off")
plt.show()
````

````python
# Bitwise Not


bitwise_not=cv.bitwise_not(rectangle)
plt.imshow(bitwise_not,cmap="gray")
plt.axis("off")
plt.show()
````

All white images into black and vice versa
