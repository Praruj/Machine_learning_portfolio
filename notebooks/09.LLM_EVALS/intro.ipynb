# LLM Evaluation

*Vibe testing:*

- It means casually testing or trying am LLM app with a few prompts and judging it by feel.
- Basically, informal, subjective and usually not repeatable prompts, which works at personal project level only.


**LLM_EVALS:**
It is systematic, repeatable test used to judge an LLm or LLM-powered system against a clear criteria. 

**Systematic:**
It shouldnot be random prompting.

**Repeatable:**
The same eval should be runnable again. If we change the prompt, model, retriever, chunking strategy or system instruction, we should be able to run the same test again and compare results. 
This is how we know wether the system involved or become worse. 

**Clear criteria:**
we must define what good means. Good answer should be:


An eval is not just a metric, it is the complete testing setup. It includes:
i. What are we evaluating ? 
ii. What does good mean ?
iii. What test cases are we using ?
iv. How are we judging the output?
v. When we are running it ?
vi. Which tool we are using ? 

We are talking about the complete process of testing an LLM powered system in a structured way. 
The goal is not just to get score but to answer practical questions like:

- Can the model be used for a particular task/application?
- Is this system good enough to ship ?
- Did prompt v2 improve over prompt v1 ?
- Is the RAG answer grounded in the retireved content ?
- Is the agent completing the task correctly ?
- Is the chatbot safe for real users?


## Model Evals

Model evals evaluate the model itself. The main idea is to test and evaluate the capabilities of a model. 
Some important capabilities include:
1. Reasoning: Can it reason ?
2. Knowledge:  Can it answer knowledge based question?
3. Maths:  Can it solve maths problem
4. Coding: Can it write code
5. Instruction following:  Can it follow instructions
6. Long Context: Can it follow long documents
7. Multimodal understanding: Can it understand image
８．Tool Use : Can it use tools


- Model capabilites are generally evaluated using benchmarks. Some famous benchmarks include: 

| Capability Area                 | Famous Benchmark(s)     | What It Checks                                                                 |
|---------------------------------|-------------------------|--------------------------------------------------------------------------------|
| General Knowledge + Reasoning   | MMLU                    | Broad subject knowledge and reasoning across domains like science, history, law, medicine, and more |
| Maths                           | GSM8K                   | Grade-school math word problems and step-by-step numerical reasoning          |
| Coding                          | HumanEval, SWE-bench    | Code generation and real-world software engineering issue solving             |
| Instruction Following           | IFEval                  | Whether the model follows explicit constraints and formatting instructions    |
| Long Context                    | Needle-in-a-Haystack    | Whether the model can find specific information hidden inside a very long context |
| Multimodal Capabilities         | MMMU                    | Reasoning over images, diagrams, charts, and visual academic problems         |
