# Open source vs Proprieatary models


Based on accessibility, control over model:

**Proprietery models:** 
It is AI models that are owned and controlled by a company. They are often refferd as black boxes because their source code, training data and learned parameters are not publicly accessible, preventing external users from fully inspecting or understanding the internal decision making process. 

It is typically accessed through paid subscriptions or usage based pricing, although limnited free access maybe provided for testing or evaluation purpose.
Thet are commonly deployed as cloud based service, where user sends requests to provider infrastructure and receive responses without direct access to the underlying model. 


**Open source:** 

Open source models are AI models whose core details are openly shared with public.
- There model architecture, weights and sometime even the training code are made public. Anyone can download them, study how they work, modify them, fine tune them for their own needs and run them on their own systems.

- Access : We can download the model files (weights) and platforms like hugging Face and run them for free though we still pay for electricity and hardware.
- Customization: Since we have specific file, we can "fine tune" the model on our own private data to make it an expert in specific field. 
- Examples: `Meta's Llama 4`, `Mistral` and `Deepseek`

Ollama is a tool that lets us to run LLM on our own computer. It makes easy to download run and manage AI models locally using very simple commands. 

Key benefits of using Ollama:

- Privacy and Data control: Our input data never leaves our infrastructure, thats crucial when data is sensitive. Eg A lawyer asks an AI to summarize a confidential case file. The document is upload to a cloud provider.

- Lower latency and offline access: No network round trip to a cloud server -> faster responses. Also works when internet is slow/ unavailable.

- Cost Predictability and potential savings: Cloud costs are often pay per request/ token and can add up for high throughput systems. Local interference is mainly upfront hardware and electricity per query can cost can be much lower at scale.

- Simple installation and setup: We install Ollama like any other app (download install and done). No complicated environment setup, Python libraries or GPU drivers needed,.
We dont need ML engieer, Anyone with laptop can get started.

- Pre built model library: Ollama has collection of ready to use models. Beginners dont need to searcg, download and configure models from multiple websites. Its like playstore for models.

- Customaization of models: We can instruction- tune the model and run it locally on our system. For our own use case.

- No vendor lock and Ip protection: You are bound to cloud vendor terms, pricing changes or usage restrictions. Our prompts are intellectual property under our control.

- Run Manage model with simple Commands: once Ollama is installed,  we can start a model by typing simple commands, no need to worry about how model will handled from where it will be downloaded. Ollama handles it for us: 
Models can be simply managed using simple commands: 
1. `Ollama pull llama2` -> download model
2. `Ollama rm llma2` -> removes model
3. `Ollama list` -> see installed models
