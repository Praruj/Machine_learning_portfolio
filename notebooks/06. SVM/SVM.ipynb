# Support Vector Machines

SVM is supervised learning method that looks at data and sorts it one of the two categories.  

## Application
1. Face Detection
2. Classification of images
3. Bioinformatics
4. Text and Hypertext categorization

## Advantages of Support Vector Machine
- High Dimensional input space.
- Sparse document vectors
- Regularization parameter


# Use Case - Implementation

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets._samples_generator import make_blobs
```

```python
X,y=make_blobs(n_samples=40, centers=2,random_state=20)
clf=svm.SVC(kernel="linear",C=1)
clf.fit(X,y)

plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=plt.cm.Paired)
```

```python
# Predict data
data=[[5,6],[3,4]]
clf.predict(data)
```

1. The first input example is predicted to belong to class 1 (often called the “positive” class in a binary problem).
2. The second input example is predicted to belong to class 0 (often the “negative” or “other” class, depending on how you encoded labels).

SVM draws a separating line (in higher dimensions, a hyperplane) between two groups of points.


```python
# plt the decision function

clf=svm.SVC(kernel="linear",C=1)
clf.fit(X,y)
plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=plt.cm.Paired)

ax=plt.gca()
xlim= ax.get_xlim()
ylim=ax.get_ylim()

xx=np.linspace(xlim[0],xlim[1],30)
yy=np.linspace(ylim[0],ylim[1],30)

YY,XX=np.meshgrid(yy,xx)
xy=np.vstack([XX.ravel(),YY.ravel()]).T
Z=clf.decision_function(xy).reshape(XX.shape)


ax.contour(XX,YY,Z,colors='k',levels=[-1,0,1], alpha=0.5, linestyles=["--","-","--"])
ax.scatter(clf.support_vectors_[:,0],clf.support_vectors_[:,1],s=100, linewidth=1,facecolors="none")
plt.show()
```

This demonstrates Support Vector Machine by seggregating two classes.
