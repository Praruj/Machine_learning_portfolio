# K-Nearest Neighbours Algorithm

It is one of simplest supervised machine learning algorithm mostly used for classification. 
   - It classifies a data point based on how its neighbours are classified. 

KNN stores all availables cases and classifies new cases based on a similarity measure. 
`K` in KNN is a parameter that refers to a number of nearest neighbours to include in the majority voting process. 

## How do we choose the Factor `K`?
The algorithm is based on feature similarity: choosing the right value of k is process called parameter tuning, and it is important for better accuracy. 
- `Sqrt(n)` where n is total number of Data points.
- odd value of k is selected to avoid confusion between two classes of data. 

### When to use KNN?
- When Data is labeled
- Noise free
- Dataset is small, since, KNN is lazy learner. 

### How does KNN algorithm work ?
1. Choose the number of neighbors `k`.
2. Compute the **distance** (usually Euclidean) between the new data point and all points in the training set.
3. Select the `k` closest neighbors.
4. Perform **majority voting** (for classification) or **averaging** (for regression) to predict the output.

### Euclidean Distance Formula

For two points \( p = (p_1, p_2, ..., p_n) \) and \( q = (q_1, q_2, ..., q_n) \):

$$
d(p, q) = \sqrt{\sum_{i=1}^{n}(p_i - q_i)^2}
$$

 

# USE CASE: Predict diabetes

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv(r"..\..\Datasets\diabetes.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Replace zero
zero_not_accepted=["Glucose","BloodPressure","SkinThickness","BMI","Insulin"]

for column in zero_not_accepted:
    df[column]=df[column].replace(0,np.nan)
    mean=int(df[column].mean(skipna=True))
    df[column]=df[column].replace(np.nan,mean)
```

- X = ALL rows (:), FIRST 8 columns (0:8)   → FEATURES for ML model (Columns 0-7)
- y = ALL rows (:), JUST column 8 (8)       → TARGET variable to predict (Column 8)

```python
# Split dataset into train and Test:

X=df.iloc[:,0:8]
y=df.iloc[:,8]

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.2)
```

### Rule of thumb:
Any algorithm that computes distances or assumes normality, scales the features. 

Without scaling: KNN = "Guess based on Glucose only"
With scaling:    KNN = "Smart guess using ALL features equally"


```python
# Feature scaling

sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
import math
math.sqrt(len(y_test))
```

```python
# Define the model
classifier=KNeighborsClassifier(n_neighbors=11,p=2,metric="euclidean")
classifier.fit(X_train,y_train)
```

```python
# Evaluate model
y_pred=classifier.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
f1_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

Fair fit in the model
