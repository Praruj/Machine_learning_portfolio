# Dense Layer

A **Dense layer** (also called a **Fully Connected layer**) is one of the most important building blocks in neural networks.

It performs a linear transformation followed by a non-linear activation function.



The dense layer computes:

$$
Z = XW + b
$$

Where:

- $X$ = Input matrix of shape $(batch\_size, input\_dim)$  
- $W$ = Weight matrix of shape $(input\_dim, output\_dim)$  
- $b$ = Bias vector of shape $(1, output\_dim)$  
- $Z$ = Linear output  

---

## 2. Activation Function

After computing $Z$, we apply a non-linear activation:

$$
A = \sigma(Z)
$$

If using the sigmoid activation:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

So the full equation becomes:

$$
A = \sigma(XW + b)
$$

---

## 3. Shape Rule for Matrix Multiplication

Matrix multiplication rule:

$$
(m \times n)(n \times p) = (m \times p)
$$

So if:

- $X$ is $(batch\_size \times input\_dim)$  
- $W$ is $(input\_dim \times output\_dim)$  

Then:

$$
Output = (batch\_size \times output\_dim)
$$

---

## 4. Single Neuron Formula

For one neuron:

$$
z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b
$$

$$
a = \sigma(z)
$$

---
A Dense layer is simply:

$$
A = activation(XW + b)
$$

```python
import tensorflow as tf

class Mydenselayer(tf.keras.layers.Layer):
    def __init__(self, input_dim, output_dim):
        super(Mydenselayer,self).__init__()

        self.w=self.add_weight((input_dim,output_dim))
        self.b=self.add_weight([1,output_dim])

    def call(self,inputs):
    #forward propagate the inputs
        z=tf.matmul(inputs,self.w)+self.b
        output=tf.math.sigmoid(z)
        return output
```
