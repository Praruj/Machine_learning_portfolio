#  Introduction to PyTorch

PyTorch is a flexible, open-source deep learning framework developed by Meta's AI Research lab (FAIR). It is widely used for its native Python integration and dynamic execution model.


##  Core Features

### 1. Tensor Computation
* **Concept:** Tensors are multi-dimensional arrays, serving as the fundamental data structure.
* **Key Advantage:** Highly compatible with **NumPy**, allowing for seamless conversion, but optimized for deep learning operations.

### 2. GPU Acceleration
* **Concept:** Offloads heavy mathematical computations from the CPU to a GPU (via NVIDIA CUDA or Apple Silicon MPS).
* **Key Advantage:** Speeds up neural network training and inference times through massive parallel processing.

### 3. Dynamic Computation Graph (Define-by-Run)
* **Concept:** PyTorch builds the computation graph *on the fly* as operations are executed, rather than pre-defining a static architecture.
* **Key Advantage:** Enables highly flexible debugging, variable-length inputs (crucial for NLP), and standard Python control flow (`if` statements, `for` loops) inside models.

### 4. Automatic Differentiation (`autograd`)
* **Concept:** The engine behind backpropagation. It automatically tracks all operations performed on tensors and calculates gradients.
* **Key Advantage:** Eliminates the need to manually derive and code partial derivatives for complex neural network layers.

### 5. Distributed Training
* **Concept:** Native tools (`torch.distributed`) designed to split workloads across multiple GPUs, machines, or cloud nodes.
* **Key Advantage:** Essential for scaling up to massive datasets and training modern Foundation Models or LLMs.

### 6. Interoperability & Ecosystem
* **Concept:** Deep integration with the broader Python data science ecosystem (NumPy, SciPy, Pandas).
* **Key Advantage:** Easily pairs with deployment tools like **ONNX** to export models to other runtimes and production environments.



| Feature | Primary Benefit |
| :--- | :--- |
| **Tensors** | Efficient multi-dimensional data handling |
| **GPU Acceleration** | Massive speedups for heavy math |
| **Dynamic Graph** | Flexible, intuitive debugging and architecture |
| **Autograd** | Hands-free gradient calculations for training |
| **Distributed** | Scalability across hardware clusters |
| **Interoperability** | Smooth integration with existing Python stacks |


##  Core PyTorch Modules 

When building models, you will constantly import these primary namespaces. Here is what they do:

| Module / Namespace | Description & Purpose | Common Use Cases |
| :--- | :--- | :--- |
| **`torch`** | The root package. Provides tensor library functions, mathematical operations, and basic creation ops. | `torch.tensor()`, `torch.zeros()`, `torch.cat()` |
| **`torch.nn`** | The Neural Network module. Contains deeply optimized building blocks (layers, activation functions, loss functions). | `nn.Linear()`, `nn.Conv2d()`, `nn.MSELoss()` |
| **`torch.nn.functional`** | Functional interface for `torch.nn`. Contains stateless operations (functions that don't hold weights). | `F.relu()`, `F.softmax()`, `F.dropout()` |
| **`torch.optim`** | A package implementing various optimization algorithms for gradient descent. | `optim.SGD()`, `optim.Adam()`, `optim.RMSprop()` |
| **`torch.autograd`** | The automatic differentiation engine that powers all neural network training. | `torch.autograd.backward()`, `torch.no_grad()` |
| **`torch.utils.data`** | Utilities for handling datasets, batching, shuffling, and multi-process data loading. | `Dataset`, `DataLoader` |
| **`torchvision` / `torchaudio`** | Official domain-specific companion libraries for computer vision and audio processing. | `torchvision.datasets`, `torchvision.transforms` |

**NOTE**

```TEXT
Accuracy → overall correctness
Precision → quality of positive predictions
Recall → quantity of positives found
F1 → balance of precision & recall

```


| Metric    | Question it answers                      |
| --------- | ---------------------------------------- |
| Accuracy  | “How often am I correct overall?”        |
| Precision | “When I say YES, how often am I right?”  |
| Recall    | “How many YES did I catch?”              |
| F1 score  | “Did I balance precision + recall well?” |
