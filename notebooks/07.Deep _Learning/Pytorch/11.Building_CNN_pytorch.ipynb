Convolutional neural networks are specialized type of neural network designed for processing grid like data like images or sequences. They work using activation functions and vatious layers. 

- Convolutional layers : data is passed through one or more convolutional layers and set of learnable filters are applied. These layers capure patterns by detecting edges, corners and texttures.

- Activation function: It is like gate that determines whether a neuron should fire or not based on weighted sum of its inputs.

- Pooling layers: Pooling layers are mainly responsible for downsampling feature maps. 

- Fully connected layers: Flattend feature maps are then passed through fully connected layers, which can process high level features for prediction or classification. 


**Applications**

1. Image classification
2. Object detection
3. Facial recognition
4. Medical imaging
5. Video analysis
