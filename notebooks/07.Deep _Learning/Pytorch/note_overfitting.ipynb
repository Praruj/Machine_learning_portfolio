## Strategic Mitigations for Overfitting

To prevent a model from memorizing noise and ensure it generalizes well to unseen data, apply one or more of the following architectural and training techniques:

* **Dataset Expansion:** Increase the volume of training data. Feeding the neural network more diverse examples directly reduces its capacity to overfit to specific patterns.
* **Regularization ($L_2$):** Append a penalty term to the loss function. In deep learning, $L_2$ regularization (weight decay) penalizes large weights, forcing the network to learn smoother, less complex decision boundaries.
* **Dropout:** Randomly deactivate a subset of neurons during each training iteration. This prevents co-adaptation of features by forcing the network to learn redundant, robust representations.
* **Data Augmentation:** Apply random transformations—such as flipping, rotating, and tilting—to input data during training. This artificially expands the dataset and is particularly effective for spatial invariance in Convolutional Neural Networks (CNNs).
* **Batch Normalization:** Normalize the inputs of each layer to stabilize and accelerate training. As a secondary benefit, the mini-batch statistics introduce slight noise, acting as a mild regularizer.
* **Early Stopping:** Monitor validation performance during training and halt the process as soon as validation loss begins to degrade, preventing the model from over-optimizing on the training set.

**Regularization ($L_2$):** 

1. Applied to Model Weights
   * Regularization penalizes large parameter values to encourage smaller, more generalizable weights.

2. Introduced via Loss Function or Optimizer
   * Method A: Via Loss Function ($L_2$ Regularization)
     Adds a penalty term directly to the original loss function ($L_0$):
     $$L = L_0 + \frac{\lambda}{2} \sum_{i} w_i^2$$
     * $L_0$: Original objective/loss function (e.g., Cross-Entropy, MSE).
     * $\lambda$: Regularization strength hyperparameter.
     * $w_i$: Individual model weights.

   * Method B: Via Optimizer (Weight Decay)
     Instead of modifying the loss, it directly modifies the gradient update rule to include $\lambda w_i$, effectively shrinking the weights during training:
     $$w_{t+1} = w_t - \eta \nabla L_0(w_t) - \eta \lambda w_t$$
     Which can be rewritten as:
     $$w_{t+1} = (1 - \eta \lambda) w_t - \eta \nabla L_0(w_t)$$
     * $w_t$: Current weight at step $t$.
     * $\eta$: Learning rate.
     * $\nabla L_0(w_t)$: Gradient of the original loss function.
     * $(1 - \eta \lambda)$: The decay factor that exponentially shrinks the weights before applying the gradient step.

3. Penalizes Large Weights
   * Encourages the network to distribute learning across multiple parameters, avoiding a heavy reliance on just a few dominant weights.

4. Reduces Overfitting
    * helps the model generalize better to unseen data by discouraging overly complex representations.

5. Controlled by Hyperparameters
    * A regularization coefficient ($\Lambda$, often set via weight_decay in optimizers) controls the strength of penalty, larger values lead to stronger regularization.

6. No effect on bias term
    * Regularization is typically applied only to weights not biases as biases dont directly.
