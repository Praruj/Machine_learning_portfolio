# Tensors in Pytorch

Tensors are data structures, `(arrays in higher dimensional array)`. It is specialized multi dimensional array designed for mathematical and computational efficiency. 
Dimension of tension refers to how many direction it is spanning. 

## Example of Tensors

1. Zero dimensional tensors : Scalars, which is not spanned in any direction.
    - Represents a single value and often used for simple metrics or constants. 
    - loss value : After a forward pass. the loss function computes a single scalar value indicatinf the difference between the predicted and actual outputs. 
    - Example : 5.0 or -3.14


2. 1- Dimensional tensors : Vectors (arrays)
    - Represents a sequence or collection of values.
    - Feature vector: In NLP, each word in a sentence maybe represented as a 1D vector using embeddings.
    - Example: [0.12,0.84,0.33] (a word embedding vector from a pre trained model like word2vec or Glove)

3. 2- Dimensional tensors (a 2D grid of numbers) : Matrices
    - Represents tabular or grid like data
    - Grayscale Image : A grayscale image can be represented as a 2d tensor, where each entry corresponds to the pixel intensity.
    - Example: [[0,255,129],[34,90,180]]

4.  Coloured Images : 3d tensors
    - Adds a third dimension, often used for stacking data.
    - RGB images : A single RGB image is represented as a 3D tensor (width * height * channels)
    - Examples RGB Image :  eg (256*256) : shape(256,256,3)


5. Batches of RGB Images : 4D Tensors
    - Batches of RGB images: A dataset of coloured images is represented as a 4D tensor (batch size * width * height*channels)
    - A batch of 32 images, each size of 128*126 with 3 color channels (RGB), would have shape [32,128,128,3]


6. Video data : 5d tensors
    - Adds a time dimension for data that changes over time . (eg video frame)
    - Video clips: Represented as a sequence of frames, where each frame is an RGB image.
    - A batch of 10 video clips, each with 16 frames of size 64*64 and 3 channels (RGB) would have shape [10,1664,64,3]

### Why tensors are useful ?

1. Mathematical Operations
    - Tensors enables efficient mathematical computations (addition, multiplication,dot product, etc) for neural network operations.

2. Representation of Real world data
    - Data like images, audio, video and text can be represented as tensors:
    - Images : Represented as 3D tensors (width * height *channels)
    - Text: Tokenized and represented as 2d or 3d tensors (sequence length * embedding size) 


3. Efficient computations
    - Tensors are optimized for hardware accelerations, allowing computations on GPUs or TPUs, which are crucial for training deep learning models.

#### Why Tensors are used in deep learning ?

1. Data storage : Training data (images,text etc) is stored in tensors.
2. Weight ans biases : The learnable parameters of nueral network (weight, biases) are stored as tensors. 
3. Matrix operations : Neural network involve operations like matrix multiplication, dor products and broadcasting - all performed using tensors.
4. Training process : During forward passes, tensors flow through the network. Gradients represented as tensors are calculated during the backward passf.

````python
# absolute (abs)
c=torch.tensor([1,3,4,5])
torch.abs(c)

# negative
d=torch.tensor([1.9,2.4,3.7,4.4])
torch.neg(d)

# round
torch.round(d)

#ceil
torch.ceil(d)       # It rounds every number in tensor up to the nearest whole integer.

# floor does the exact opposite of ceil. It forcefully rounds every decimal number down to the nearest whole integer, dragging it down to the "floor."
torch.floor(d)

#clamp
torch.clamp(d,min=2.0,max=4.0) # setting physical boundaries (a floor and a ceiling) for values.
````

```Text
Exploding gradients can cause numbers to become violently large during training, breaking the neural network. We use `torch.clamp()` to put safety rails on the data, forcing wild values to stay within a stable, readable range like `(-1.0 or 1.0)`.
```

# Vector matrix and Tensor

The only thing that makes the Matrix "different" from the Vector is a tiny piece of metadata metadata called the **stride**. The stride tells the system: *"To move down one row, skip 3 slots in memory."* By changing the stride metadata, a computer can instantly view a Vector as a Matrix, or a Matrix as a 3D Tensor (`tensor.view()` or `tensor.reshape()`) **without moving or changing a single byte of the actual data.**

### Summary
They have different names because human language developed hierarchically as we discovered higher dimensions of geometry. But logically, they are a continuum. A vector is just a matrix with a width of 1; a matrix is just a 3D tensor with a depth of 1. It is Tensors all the way down.
