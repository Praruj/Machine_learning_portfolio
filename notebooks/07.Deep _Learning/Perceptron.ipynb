# Perceptron

- A perceptron is the basic building block of a neural network. It is a supervised machine learning algorithm designed for binary classification tasks. The perceptron forms the foundation of deep learning models by learning patterns from input data and making predictions based on weighted inputs.

- It is also called `Mathematical model` or `functions`.  

## Structure of a Perceptron

```text
Input Features → Weights → Summation → Activation Function → Output
```

---

## Mathematical Representation

A perceptron calculates the weighted sum of inputs:


$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$


Where:

- \(x_1, x_2, ..., x_n\) = Input features  
- \(w_1, w_2, ..., w_n\) = Weights  
- \(b\) = Bias term  
- \(z\) = Linear combination of inputs  

---

## Activation Function

The output is generated using an activation function:

## Activation Function

The output is generated using an activation function:

$$
y = f(z)
$$

For a basic perceptron:

$$
f(z)=
\begin{cases}
1 & \text{if } z \geq 0 \\
0 & \text{if } z < 0
\end{cases}
$$

---

## Working Process

1. Take input values
2. Multiply inputs by weights
3. Add bias
4. Apply activation function
5. Produce output

---

## Simple Diagram

```text
 x1 ----(w1)---\
 x2 ----(w2)----> [ Σ ] ---> Activation ---> Output
 x3 ----(w3)---/
                  +
                 Bias
```
---

### Training the Perceptron

During the training process, the perceptron learns the optimal values of:

- \(w_1, w_2, ..., w_n\) → Weights  
- \(b\) → Bias  

These values are adjusted repeatedly using training data to minimize prediction errors.

After training, the learned weights and bias are used to make predictions on new input data.

The prediction is calculated as:

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

Then the activation function determines the final output:

$$
y = f(z)
$$

## Geometric Intuition of a Perceptron across Dimensions

The decision boundary of a perceptron is always **one dimension lower ($N-1$)** than the dimension of the feature space ($N$) it is trying to separate.

### Dimension Breakdown

* **1D Space (1 Feature):** The perceptron acts as a **Point**
  * *Visual:* A single dot on a number line separating values to the left and right.
* **2D Space (2 Features):** The perceptron acts as a **Line**
  * *Visual:* A straight line cutting a flat 2D scatter plot into two halves.
* **3D Space (3 Features):** The perceptron acts as a **Plane**
  * *Visual:* A flat sheet of paper slicing through a 3D room of data points.
* **4D+ Space ($N$ Features):** The perceptron acts as a **Hyperplane**
  * *Visual:* An $(N-1)$ dimensional flat divider. While impossible to mentally visualize, it follows the exact same mathematical properties as a line or plane.

---

### Summary Table

| Data Dimensions ($N$) | Feature Space Type | Perceptron Boundary ($N-1$) |
| :--- | :--- | :--- |
| **1D** | Linear Line | **Point** |
| **2D** | Flat Surface ($X, Y$) | **Line** |
| **3D** | Volume ($X, Y, Z$) | **Plane** |
| **4D and higher** | Hyperspace | **Hyperplane** |

> 💡 **Core Takeaway:** Because a perceptron can *only* create flat, straight boundaries (points, lines, planes, and hyperplanes), it is strictly limited to solving **linearly separable** datasets.

## Key Differences: Neuron vs. Perceptron

### 1. Complexity
* **Biological Neuron:** Incredibly complex; uses intricate electrochemical signaling and contains thousands of synaptic connections per cell.
* **Perceptron:** Highly simplified; a basic mathematical model designed to mimic only the fundamental "input-threshold-output" behavior.

### 2. Processing
* **Biological Neuron:** Processes information asynchronously using a mix of analog (chemical gradients) and digital-like (electrical action potentials) signals.
* **Perceptron:** Processes information synchronously using purely digital and mathematical operations (multiplying inputs by weights, adding a bias, and applying an activation function).

### 3. Neuroplasticity
* **Biological Neuron:** Features true neuroplasticity, meaning the brain can physically grow, prune, and reorganize its structural connections based on experience.
* **Perceptron:** Lacks structural plasticity; it "learns" solely by adjusting static numerical weights during training while its architecture remains completely rigid.

# Why We Use StandardScaler in Perceptron

## Problem
The features in the dataset have different ranges:

- CGPA → values around 5–9
- IQ → values around 80–230

Machine learning algorithms like Perceptron are sensitive to feature scale.

Without scaling:
- Larger values dominate learning
- Model becomes biased
- Decision boundary may be incorrect

---

## Solution: StandardScaler

`StandardScaler` transforms data so all features are on a similar scale.

Formula:

```math
z = \frac{x - \mu}{\sigma}
```

Where:
- `μ` = mean
- `σ` = standard deviation


After scaling:
- Mean becomes 0
- Standard deviation becomes 1

---

## Benefits
- Faster convergence
- Better model performance
- Fair contribution from all features
- Proper decision boundary visualization

---

## Code

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

````python
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# features and target
x = df.iloc[:, 1:3].values
y = df.iloc[:, -1].values

# scaling
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# model
p = Perceptron()
p.fit(x_scaled, y)

# plot
plot_decision_regions(x_scaled, y, clf=p)

plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.title("Perceptron Decision Boundary")

plt.show()
````

> “higher CGPA + higher IQ → more placement chance”

 Both features are treated equally  
 Boundary is stable and meaningful

## 4. Convergence 

### What is convergence?

Convergence means:

> The algorithm stops updating weights because it has found a correct separating line.

### When does perceptron converge?

It converges only if:

Data is **linearly separable**  
There exists at least one perfect separating line  

This is known as the:

> **Perceptron Convergence Theorem**

### What happens during training?

- At first → many mistakes → frequent updates  
- Later → fewer mistakes → smaller updates  
- Finally → no mistakes → training stops  

### Important note

If data is **not linearly separable**:  
- Perceptron will never fully converge  
- It will keep oscillating (updating forever)

## 6. Adding Decision Regions (Positive & Negative)

You can visualize positive and negative regions using the `mlxtend` library:


### How it works:

- **Background color:** Each color represents a region predicted as one class.  
  - e.g., Blue → class 0 (negative region)  
  - Orange → class 1 (positive region)  
- **Data points:** The scatter points show actual training samples.  
- **Decision boundary:** The border between colors is the line separating positive and negative regions.


### Key Points:

- Always scale features before plotting for consistent regions.  
- The positive region corresponds to the class predicted as 1.  
- The negative region corresponds to the class predicted as 0.  
- You can interpret these regions just like your Perceptron equation:

### Positive and Negative Regions (Equation Form)

- **Positive region (class 1):**

$$
w_1 x_1 + w_2 x_2 + b > 0
$$

- **Negative region (class 0):**

$$
w_1 x_1 + w_2 x_2 + b < 0
$$

Where:
- `x1` = first feature (CGPA)  
- `x2` = second feature (IQ)  
- `w1, w2` = learned weights  
- `b` = bias  

**Interpretation:**
- Points in the **positive region** → predicted as class 1  
- Points in the **negative region** → predicted as class 0  
- Points on the line (`= 0`) → the decision boundary

# Perceptron Notes 

## 1. Perceptron Trick
- Linear classifier separates data with a line:
  
$$
w_1 x_1 + w_2 x_2 + b = 0
$$

- Start with random weights → update when prediction is wrong:

$$
w = w + yx,\quad b = b + y
$$`

- Converges if data is **linearly separable**  
- Intuition: move line toward correct class, repeat until stable

---

## 2. Labeling Positive/Negative Regions

- **Positive region (class 1):**  

$$
w_1 x_1 + w_2 x_2 + b > 0
$$

- **Negative region (class 0):**  

$$
w_1 x_1 + w_2 x_2 + b < 0
$$

- Points on the line (`=0`) → decision boundary  
- Regions can be visualized using `mlxtend.plotting.plot_decision_regions`

---

## 3. Transformations

- Scale features to similar range → faster convergence
  ```python
  from sklearn.preprocessing import StandardScaler
  x_scaled = StandardScaler().fit_transform(x)
  ```
- Optional: rotate, shift, or normalize features for better separation  
- Scaling ensures all features contribute equally to the boundary


### Key Transformation Rule

When transforming points (scaling, rotating, translating):

- Use **`+`** → move toward **positive region**  
- Use **`-`** → move toward **negative region**

**Intuition:**  
The vector \([w_1, w_2]\) points “forward” toward the positive region.  
- Moving **along the vector** → positive region  
- Moving **against the vector** → negative region

### Key Idea

- **Learning rate (`η`)** controls **how big the weight updates are**.  
- Typical value: `η = 0.01`  

---

### Weight Update Rule

For a misclassified point:

$$
w_{\text{new}} = w_{\text{old}} + \eta \cdot y \cdot x
$$

where:
- $x$ = feature vector (coordinates)  
- $y$ = true label (1 or 0/-1)  
- $\eta$ = learning rate  


---

### Intuition

1. Multiply **coordinates by learning rate** → small step toward correct classification.  
2. Update coefficients:

$$
\text{new coefficient} = \text{old coefficient} + (\eta \cdot y \cdot x)
$$

- Positive region → increase weights along the point direction  
- Negative region → decrease weights along the point direction  

> Small learning rate = slow but stable convergence  
> Large learning rate = fast but may overshoot
